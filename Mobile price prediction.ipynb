```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,Y,test_size=0.2,random_state=7)
```

```python
def get_score(model,train_x,test_x,train_y,test_y):
    model.fit(train_x,train_y)
    return model.score(test_x,test_y)
```

```python
get_score(RandomForestClassifier(),train_x,test_x,train_y,test_y)
```

```python
get_score(GaussianNB(),train_x,test_x,train_y,test_y)
```

```python
get_score(KNeighborsClassifier(),train_x,test_x,train_y,test_y)
```

```python
get_score(svm.SVC(),train_x,test_x,train_y,test_y)
```
